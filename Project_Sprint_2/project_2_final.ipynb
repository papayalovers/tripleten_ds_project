```python
# Menghitung nasabah yang memiliki anak 20
children_20 = len(df_cre[df_cre['children'] == 20])
childre_min1 = len(df_cre[df_cre['children'] == -1])
print(f"Jumlah nasabah yang memiliki data anak 20: {children_20} nasabah")
print(f"Jumlah nasabah yang memiliki data anak -1: {childre_min1} nasabah")
print(f"Persentase nilai bermasalah dari keseluruhan data: {((children_20 + childre_min1) / len(df_cre)):.3%}")
```

```python
#memeriksa mean dan median sebelum menggannti nilai
child_mean_med = df_cre.groupby('gender').agg(
{
    'children' : ['mean','median']
})

child_mean_med 
```

Dilihat dari persentase dibilang cukup rendah, dan juga dilihat dari rata-rata dan median dari children tidak mungkin anak dari nasabah merupakan bilangan ganjil(float) serta dengan mempertimbangkan bahwa outlier dari children itu sangat signifikan maka data akan diganti dengan median.

```python
# memperbaiki data dengan median 
df_cre['children'] = df_cre['children'].replace([20,-1], child_mean_med['children']['median']['F'])
```

Karena usia 0 tahun terdapat lebih banyak pada gender wanita. Maka nilai 0 akan digantikan dengan usia rata-rata wanita.

### 4.1 Memperbaiki nilai yang hilang di `total_income` <a id='total_income'></a>

Mulai dengan mengatasi total nilai pendapatan yang hilang. Buat kategori usia untuk nasabah. Buat kolom baru yang memuat kategori usia. Strategi ini dapat membantu untuk menghitung total nilai pendapatan.

```python
# Mari kita tulis sebuah fungsi untuk menghitung kategori usia
def dob_grouped(row):
    age = row['dob_years']
    if age < 26:
        return '15-25'
    elif age < 46:
        return '26-45'
    elif age < 66:
        return '46-65'
    else:
        return '65+'
    
```

```python
#  Tulis fungsi yang akan kita gunakan untuk mengisi nilai yang hilang

def calc_income(row):
    income_t = row['income_type']
    #age = row['age_grouped']
    if age == '15-25':
        if income_t == 'business':
            return income_age_mean_grouped['total_income']['15-25']['business']
        elif income_t == 'civil servant':
            return income_age_mean_grouped['total_income']['15-25']['civil servant']
        elif income_t == 'employee':
            return income_age_mean_grouped['total_income']['15-25']['employee']
        elif income_t == 'retiree':
            return income_age_mean_grouped['total_income']['15-25']['retiree']
        else:
            return income_age_mean_grouped['total_income']['15-25']['student']
    elif age == '26-45':
        if income_t == 'business':
            return income_age_mean_grouped['total_income']['26-45']['business']
        elif income_t == 'civil servant':
            return income_age_mean_grouped['total_income']['26-45']['civil servant']
        elif income_t == 'employee':
            return income_age_mean_grouped['total_income']['26-45']['employee']
        elif income_t == 'entrepreneur':
            return income_age_mean_grouped['total_income']['26-45']['entrepreneur']
        elif income_t == 'paternity / maternity leave':
            return income_age_mean_grouped['total_income']['26-45']['paternity / maternity leave']
        elif income_t == 'retiree':
            return income_age_mean_grouped['total_income']['26-45']['retiree']                            
        else:
            return income_age_mean_grouped['total_income']['26-45']['unemployed']
    elif age == '46-65':
        if income_t == 'business':
            return income_age_mean_grouped['total_income']['46-65']['business']
        elif income_t == 'civil servant':
            return income_age_mean_grouped['total_income']['46-65']['civil servant']
        elif income_t == 'employee':
            return income_age_mean_grouped['total_income']['46-65']['employee']
        else:
            return income_age_mean_grouped['total_income']['46-65']['retiree']     
    elif age == '65+':
        if income_t == 'business':
            return income_age_mean_grouped['total_income']['65+']['business']
        elif income_t == 'civil servant':
            return income_age_mean_grouped['total_income']['65+']['civil servant']
        elif income_t == 'employee':
            return income_age_mean_grouped['total_income']['65+']['employee']
        else:
            return income_age_mean_grouped['total_income']['65+']['retiree']   
    else:
        return False
```
