### Mengganti nilai yang hilang pada kolom `model_year` <a id='model_year'></a>

Pertama filter tabel pada kolom tersebut tanpa nilai yang hilang lalu lihat deskripsinya.

````python
# membuat tabel data tanpa nilai yang hilang di kolom model_year sebagai acuan mengisi nilai yang hilang
model_year_no_missing = data.query('model_year.notna()')

model_year_no_missing['model_year'].describe()
````

````python
model_year_no_missing.head()
````

Dapat dilihat terdapat jenis tahun kendaraan yang terpaut jauh dibawah pada tahun 1908 yang sepertinya angka ini adalah outlier atau data yang tidak wajar. Hal ini mengakibatkan rata-rata dari kolom `model_year` menjadi turun atau lebih rendah dari median. Jadi untuk mengisi nilai yang hilang pada kolom ini akan dilakukan dengan mencari median dari tahun kendaraan (`model_year`) untuk setiap model kendaraan (`model`) berdasarkan tabel yang telah difilter tanpa nilai yang hilang. 

````python
# mendapatkan median tahun kendaraan berdasarkan model kendaraan
model_median = model_year_no_missing.groupby('model').agg({
    'model_year' : 'median'
})
model_median
````

````python
# Mengecek distribusi nilai yang hilang pada model_year
model_year_missing = data.query('model_year.isna()')
model_year_missing.head()
````

````python
# menghitung persentase nilai yang hilang pada kolom model_year terhadap seluruh dataset
percent_model_year = f"Persentase terhadap seluruh dataset = {(len(model_year_missing) / len(data)):.2%}"
percent_model_year
````

````python
#melihat nilai pada dataset model_median sebagai acuan membuat fungsi
model_median['model_year']['ford f-150']
````

````python
#membuat fungsi dengan median berdasarkan jenis kendaraan
def model_year_cal(row):
    i = row['model']
    return model_median._get_value(i, 'model_year')
````

````python
#melihat data pada baris/index pertama pada nilai yang hilang
model_year_missing.loc[1]
````

Perhatikan model pada contoh data yaitu `ford f-150` dan perhatikan nilai median pada `ford f-150` diatasnya lagi. Lalu kemudian uji fungsi apakah bekerja dengan baik atau tidak.

````python
#uji coba apakah hasil fungsi apakah sesuai
model_year_cal(model_year_missing.loc[1])
````

Hasilnya sesuai maka lanjut terapkan fungsi diseluruh dataset. Apakah hasilnya sesuai yang diharapkan ? Jebrett......

````python
#menerapkan fungsi ke seluruh dataset
data['model_year'] = data['model_year'].fillna(data.apply(model_year_cal, axis=1))
````

- rata-rata berdasarkan jenis dan tahun kendaraan.
- rata-rata berdasarkan tahun kendaraan.

Saya mengambil keputusan seperti ini agar tidak terjadi perubahan yang signifikan disaat analisis tahap lanjut. Tapi saya menghadapi satu kendala dalam membuat fungsi karena terdapat nilai tahun `1929` yang tidak ada satupun patokannya didalam dataset.
````python
#membuat fungsi
def odometer_cal(row):
    t = row['model']
    y = row['model_year']
    try:
        return type_med_mean._get_value(index=(t,y), col=('odometer','mean')) # mengisi nilai rata-rata berdasarkan dengan jenis dan tahun kendaraan
    except:
        try:
            return i._get_value(y) # mengisi nilai dengan rata-rata berdasarkan tahun
        except:
            return 0
````

Berdasarkan informasi dari sebelumnya terdapat 3 kolom yang berisi outlier yaitu:

Lihatlah histogram yang kamu buat diatas, range nilai pada sumbu x terlihat sangat jauh. Hal ini memperlihatkan bahwa kemungkinan percilan atau outlier berada pada rentang nilai diatas nilai Q3 dan upper limit dari boxplot. Dan itu terbukti dalam boxplot yang kamu tampilkan diatas.

````python
#membuat fungsi untuk menghapus outlier
def limit_func(data, col):
    #print(col)
    Q1 = data[col].quantile(.25)
    Q3 = data[col].quantile(.75)
    IQR = Q3 - Q1
    upper_limit = Q3+1.5*IQR
    lower_limit = Q1-1.5*IQR   
    data = data[(data[col] >= lower_limit) & (data[col] <= upper_limit) ]
        
    return data
````

<div class="alert alert-success">
<b>Chamdani's comment v.1</b> <a class="tocSkip"></a>

Hebat, Kerja bagus!

</div>

````python
# membuat dataset baru dengan nama good_data
good_data = data.copy()

# melakukan looping sampai dataset baru benar-benar bersih dari outlier
for col in ['price','model_age','odometer_mean_year']:
    data_fil = limit_func(good_data, col)
    while len(data_fil) != len(good_data):
        good_data = data_fil.copy()
        data_fil = limit_func(good_data,col)
````

````python
data_1 = data.copy()

for column in ['price','model_age','odometer_mean_year']:
    while True:
        clean_data = limit_func(data_1, column)
        if len(clean_data) == len(data_1):
            break
        data_1 = clean_data
````
