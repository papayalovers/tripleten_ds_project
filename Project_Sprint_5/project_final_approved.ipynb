Ternyata ada nama games yang tahun rilisnya sama walaupun berbeda platform dan ada juga nama games yang tahun rilisnya berbeda ditiap platform. Maka dari itu kita disini mungkin akan mengisi nilai yang hilang pada kolom `year_of_release` hanya dengan median dari `year_of_release` untuk tiap platform.

```python
# membuat fungsi untuk mengisi `year_of_release`
def year_release_input(data):
    platform =  data['platform']
    return df_games.query('platform == @platform')['year_of_release'].median() # jika tidak, maka akan di isi dengan median berdasarkan platform
```

Kita disini mungkin berasumsi bahwa terdapat nama yang sama diantara rating dengan nilai yang hilang dengan rating tanpa nilai yang hilang. Untuk itu mari kita simpan di 2 variabel dan memeriksa apakah terdapat nama yang sama.

```python
# menyimpan data dengan rating missing
rating_missing = df_games.query('rating.isna()')
# menyimpan data dengan rating not missing
rating_no_missing = df_games.query('rating.notna()')
```

```python
# melihat data untuk mencari nama games yang sama di antara rating missing dan rating not missing
rating_no_missing.query('name in @rating_missing["name"]').head()
```

```python
# melihat distribusi keseluruhan rating dengan nama games `Call of Duty: Ghosts' dan 'The Elder Scrolls V: Skyrim'
df_games.query('name in ["Call of Duty: Ghosts","The Elder Scrolls V: Skyrim"] ')
```

Seperti dugaan kita sebelumnya, terdapat nama yang sama diantara rating dengan nilai yang hilang dan tidak. Untuk itu kita akan membuat fungsi untuk mengisi nilai yang hilang berdasarkan nama gamesnya, jika terdapat nama yang tidak sama kita akan mengisinya berdasarkan rating terbanyak berdasarkan genre.

```python
# menyimpan rating di dalam variabel yang akan digunakan didalam fungsi
rating_value = rating_no_missing[['name','rating']].query('name in @rating_missing["name"]').drop_duplicates(subset='name').set_index('name')
rating_value.head()
```

```python
# membuat fungsinya 
def rating_input(data):
    name = data['name']
    genre = data['genre']
    try:
        return rating_value.loc[name, 'rating'] # jika nama terdapat di rating_value maka rating akan diambil
    except:
        return rating_no_missing.query('genre == @genre')['rating'].mode()[0] # jika tidak maka diisi dengan rating terbanyak berdasarkan genre
```

```python
# menerapkan fungsi dengan mengisi nilai yang hilang
df_games['rating'] = df_games['rating'].fillna(df_games.apply(rating_input, axis=1))
```

```python
#memeriksa kembali distribusi dikolom rating
df_games.rating.value_counts(dropna=False)
```

Hasilnya sudah sesuai yang diharapkan dan kita sudah selesai mengatasi nilai yang hilang disetiap kolom. Langkah selanjutnya kita akan mengubah tipe data yang sesuai untuk setiap kolomnya.

## Analisis dan Eksploratif Data

```python
# fungsi untuk dekorasi plot
def bar_height_visualization(data):
    for bar in data.patches:
        data.annotate(format(bar.get_height().astype('int')),
                                     (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                                     ha = 'center', va = 'center',
                                     size = 10,
                                     xytext = (0,25),
                                     textcoords = 'offset points',
                                     rotation = 60)
        
```

Mari kita lihat jumlah game rilis tiap tahunnya dari keseluruhan dataset yang kita punya.

```python
# melihat distribusi games yang dirilis pada tahun yang berbeda
plt.figure(figsize=(15,6))
year_release_plot = sns.countplot(data=df_games, x='year_of_release', palette='rocket')

# memanggil fungsi untuk mendapatkan visualisasi angka diatas plot
bar_height_visualization(year_release_plot)

# fitur plot
plt.title('Jumlah Game Rilis Tiap Tahunnya', size=20)
plt.xlabel('Tahun', size= 14)
plt.ylabel('Jumlah', size= 14)
plt.ylim(0, 1700)
plt.xticks(rotation = 60, size=12)
plt.show()
```

Ternyata games rilis terbanyak terhitung pada tahun `2008` dan `2009`. Kemudian produktifitas games rilis menurun hingga tahun saat ini (laporan diterima). Sekarang mati kita lihat penjualan terbanyak dari setiap platform yang ada didataset kita.

### Mengkategorikan Usia Tiap Platform

Selanjutnya kita akan mengkategorikan usia platform menjadi beberapa kelompok, hal ini bertujuan untuk dapat menghitung jumlah platform dalam rentang usia tertentu.

```python
# fungsi ini untuk mengkategorikan usia platform
def age_grouped(data):
    age = data['platform_age']
    if age <= 5:
        return "0-5"
    elif age <= 10:
        return "6-10"
    elif age <= 15:
        return "11-15"
    else:
        return "15++"
```

```python
# menerapkan fungsi dan membuat kolom baru
platform_sales_information['age_grouped'] = platform_sales_information.apply(age_grouped, axis=1)
```

```python
# melihat hasil setelah fungsi diterapkan
platform_sales_information.sample(5)
```

```python
plt.figure(figsize=(10, 8)) #ukuran plot
# visualisasi menggunakan scatterplot untuk kategori usia setiap platform
age_platform_plot = sns.scatterplot(data=platform_sales_information.sort_values('year_of_release'), x='year_of_release', y='platform', hue='age_grouped', palette='husl')

# fitur plot
plt.title("Usia Platform", size=20)
plt.ylabel("Tahun Rilis", size=14)
plt.xlabel("Platform", size=14)
plt.grid(True)
plt.show()
```

**Kesimpulan**

Pada plot diatas dapat disimpulkan bahwa:
* Platform baru muncul biasanya rentang 0 sampai 3 tahun setelah platform sebelumnya rilis.
* Usia platform pada umumnya berkisar dibawah 10 tahun yang mana pada akhirnya popularitas platform tersebut memudar dan tidak mendapatkan penjualan apapun setelahnya.

## Uji Hipotesis

### Menguji Hipotesis Pertama

Kita akan menentukan h0 dan alternatifnya

- h0 = Rata-rata rating pengguna platform Xbox One dan PC adalah **sama**.
- h1 = Rata-rata rating pengguna platform Xbox One dan PC adalah **berbeda**.

```python
# memfilter platform xbox
xbox_platform = df_games.query('platform == "XOne"')
```

```python
# menghitung varians dari platform xbox
np.var(xbox_platform.user_score)
```

```python
# memfilter platform pc
pc_platform = df_games.query('platform == "PC"')
```

```python
# menghitung varians dari platform pc
np.var(pc_platform.user_score)
```

Untuk variansnya kurang lebih sama maka kita akan mengatur equal_var menjadi True.

```python
#pengujian hipotesis
alpha = 0.05 # alpha 5%

results = st.ttest_ind(xbox_platform.user_score, pc_platform.user_score, equal_var=True) 

print('p-value:', results.pvalue) 

if (results.pvalue < alpha):
    print("Kita menolak hipotesis nol dan menerima hipotesis alternatif")
else:
    print("Kita tidak bisa menolak hipotesis nol")
```

**Kesimpulan**

Rata-rata rating pengguna platform Xbox One dan PC adalah **berbeda**.

### Menguji Hipotesis Kedua

Kita akan menentukan h0 dan alternatifnya

- h0 = Rata-rata rating pengguna genre Action dan Sports **sama**.
- h1 = Rata-rata rating pengguna genre Action dan Sports **berbeda**.

```python
# memfilter genre action
action_rating = df_games.query('genre == "Action"')
```

```python
# menghitung varians action
np.var(action_rating.user_score)
```

```python
# memfilter genre sport
sport_rating = df_games.query('genre == "Sports"')
```

```python
# menghitung varians sport
np.var(sport_rating.user_score)
```

Ternyata variansnya sama maka kita akan mengatur equal_var menjadi True dan menetapkan alpha sebesar 5%.

```python
#pengujian hipotesis
alpha = 0.05 # alpha 5%

results = st.ttest_ind(action_rating.user_score, sport_rating.user_score, equal_var=True) 

print('p-value:', results.pvalue) 

if (results.pvalue < alpha):
    print("Kita menolak hipotesis nol dan menerima hipotesis alternatif")
else:
    print("Kita tidak bisa menolak hipotesis nol")
```

**Kesimpulan**

Rata-rata rating pengguna genre Action dan Sports **berbeda**.
